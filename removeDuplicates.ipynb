```python
import pandas as pd
```

```python
df_pub = pd.read_excel("sheet_test.xlsx")
df_gSchol =  pd.read_excel("Sheet_test_scholar_scraper.xlsx")
```

```python
len(df_pub)
```

```python
len(df_gSchol)
```

```python


df_pub_selected = df_pub[['title', 'abstract', 'journal', 'authors']]
df_gSchol_selected = df_gSchol[['Title', 'Abstract', 'Journal', 'Authors']]

# Renaming columns in df2 to match the columns in df1
# 
df_gSchol_selected = df_gSchol_selected.rename(columns={'Title': 'title', 'Abstract':'abstract', 'Journal':'journal', 'Authors':'authors'})
df_gSchol_selected.columns = ['title', 'abstract', 'journal', 'authors']
# Performing the merge based on the selected columns
merged_df = pd.merge(df_pub_selected, df_gSchol_selected, on=['title', 'abstract', 'journal', 'authors'], how='outer')

# Displaying the resulting merged data frame
merged_df
```

```python
merged_df['title']
```

```python
merged_df = merged_df.apply(lambda x: x.str.lower() if x.dtype == 'O' else x)
# Strip leading/trailing whitespaces
merged_df = merged_df.apply(lambda x: x.str.lstrip() if x.dtype == 'O' else x)
merged_df = merged_df.apply(lambda x: x.str.rstrip() if x.dtype == 'O' else x)
merged_df
```

```python
#retrieving the duplicated records
df_duplicates = merged_df[merged_df.duplicated('title')]
```

```python
df_duplicates
```

```python
df = merged_df.drop_duplicates('title')
```
